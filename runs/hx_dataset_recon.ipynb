```python
from pathlib import Path
```

```python
hx_sf_label = Path("/mnt/ssd/hbli/datasets/songformer/songformdb/HX/labels/")
hx_origin_label = Path("/mnt/ssd/hbli/datasets/harmonixset/segments/")
```

```python
hx_sf_files = sorted(hx_sf_label.rglob("*.txt"))
hx_origin_files = sorted(hx_origin_label.rglob("*.txt"))
```

```python
def files_equal_text(a, b, encoding="utf-8"):
    with a.open("r", encoding=encoding, errors="replace") as fa, b.open("r", encoding=encoding, errors="replace") as fb:
        while True:
            la = fa.readline()
            la = la.strip()
            lb = fb.readline()
            lb = lb.strip()
            if la != lb:
                return False
            if la == "":
                return True
```

```python
mismatch = []
for hx_sf_fp in hx_sf_files:
    hx_origin_fp = Path("/mnt/ssd/hbli/datasets/harmonixset/segments/" + hx_sf_fp.name[3:])
    if files_equal_text(hx_sf_fp, hx_origin_fp):
        mismatch.append(hx_sf_fp)
```
